```python
from matplotlib import pyplot as plt
```

```python
years = [2005, 2010, 2015, 2016, 2017, 2018]
```

```python
# Data obtained at the following link: http://www.fao.org/state-of-food-security-nutrition/en/
world = [14.5, 11.8, 10.6, 10.7, 10.8, 10.8]
```

```python
Africa = [21.2, 19.1, 18.3, 19.2, 19.8, 19.9]
Northern_Africa = [6.2, 5.0, 6.9, 7.0, 7.0, 7.1]
Sub_Saharan_Africa = [24.3, 21.7, 20.9, 22.0, 22.7, 22.8]
Eastern_Africa = [34.3, 31.2, 29.9, 31.0, 30.8, 30.8]
Middle_Africa = [32.4, 27.8, 24.7, 25.9, 26.4, 26.5]
Southern_Africa = [6.5, 7.1, 7.8, 8.5, 8.3, 8.0]
Western_Africa = [12.3, 10.4, 11.4, 12.4, 14.4, 14.7]
Asia = [17.4, 13.6, 11.7, 11.5, 11.4, 11.3]
Central_Asia = [11.1, 7.3, 5.5, 5.5, 5.7, 5.7]
Eastern_Asia = [14.1, 11.2, 8.4, 8.4, 8.4, 8.3]
South_eastern_Asia = [18.5, 12.7, 9.8, 9.6, 9.4, 9.2]
Southern_Asia = [21.5, 17.2, 15.7, 15.1, 14.8, 14.7]
Western_Asia = [9.4, 8.6, 11.2, 11.6, 12.2, 12.4]
Western_Asia_and_Northern_Africa = [8.0, 7.1, 9.2, 9.5, 9.8, 9.9]
LATIN_AMERICA_AND_THE_CARIBBEAN = [9.1, 6.8, 6.2, 6.3, 6.5, 6.5]
Caribbean = [23.3, 19.8, 18.3, 18.0, 18.0, 18.4]
Latin_America = [8.1, 5.9, 5.3, 5.5, 5.7, 5.7]
Central_America = [8.4, 7.2, 6.3, 6.1, 6.1, 6.1]
South_America = [7.9, 5.3, 4.9, 5.3, 5.5, 5.5]
OCEANIA = [5.5, 5.2, 5.9, 6.0, 6.1, 6.2]
```

```python
plt.figure(figsize=(13,10))
plt.plot (years, world)
plt.plot (years, Africa)
plt.plot (years, Asia)
plt.plot (years, LATIN_AMERICA_AND_THE_CARIBBEAN)
plt.plot (years, OCEANIA)
plt.xlabel('Year')
plt.ylabel('Prevalence of undernourishment (%)')
plt.legend(['world', 'Africa', 'Asia', 'Latin America and Caribbean', 'Oceania'],  loc='center right')
plt.show()
```

```python
plt.figure(figsize=(13,10))
plt.plot (years, Africa)
plt.plot (years, Northern_Africa)
plt.plot (years, Sub_Saharan_Africa)
plt.plot (years, Eastern_Africa)
plt.plot (years, Middle_Africa)
plt.plot (years, Southern_Africa)
plt.plot (years, Western_Africa)
plt.xlabel('Year')
plt.ylabel('Prevalence of undernourishment (%)')
plt.legend(['Africa', 'Northern Africa', 'Sub Saharan Africa', 'Eastern Africa', 'Middle Africa', 
            'Southern Africa', 'Western Africa'],  loc='center left')
plt.show()
```

```python
plt.figure(figsize=(13,10))
plt.plot (years, Asia)
plt.plot (years, Central_Asia)
plt.plot (years, Eastern_Asia)
plt.plot (years, South_eastern_Asia)
plt.plot (years, Southern_Asia)
plt.plot (years, Western_Asia)
plt.plot (years, Western_Asia_and_Northern_Africa)
plt.xlabel('Year')
plt.ylabel('Prevalence of undernourishment (%)')
plt.legend(['Asia', 'Central Asia', 'Eastern Asia', 'South Eastern Asia', 'Southern Asia', 
            'Western Asia', 'Western Asia and Northern Africa'],  loc='upper right')
plt.show()
```
